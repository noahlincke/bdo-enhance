```python
import pandas as pd
import numpy as np
```

```python
penPrice=16.9e9
tetPrice=1.65e9
triPrice=755000000
memoryFragPrice=1330000
concbsPrice=2700000
marketplaceTax=0.845

perFailStack=0.03
```

```python
#Actual Realtime Simulator!
#See your chance of this many fails as you go!

print("Input starting stack:")
stack=int(input())
print("Do you want to enhance automatically till success or do it yourself? Answer either ['auto','self']")
autoFlag=input()
print("Loading up Pen Dande with a",stack,"stack")
print("Type 'go' to enhance, or 'exit' to quit")
userIn=input()
tries=0
runningCost=0
cumulativeProbability=1
while (userIn!='exit'):
    tries+=1
    runningCost=runningCost+concbsPrice
    print("Try #", tries)
    rand=np.random.random()
    print("rand",rand)
    chance=0.01*((0.03000*stack)+0.30) #non-pen this is not always true
    print("chance",chance)
    upgraded=rand<chance
    if (upgraded):
        if (runningCost>=1e9):
            print("\n\nGot it after",round(runningCost/1e9,4),"billion silver")
        else:
            print("\n\nGot it after",round(runningCost/1e6,4),"million silver")
        break;
    else:
        stack+=6
        runningCost=runningCost+((tetPrice-(marketplaceTax*triPrice))+(5*memoryFragPrice))
        print("Failed, but stack is now",stack)
        if (runningCost>=1e9):
            print("Currently",round(runningCost/1e9,4),"billion in the hole")
        else:
            print("Currently",round(runningCost/1e6,4),"million in the hole")
        cumulativeProbability=cumulativeProbability*(1-chance)
        if (cumulativeProbability<0.1):
            print("Wow! You are in the top 10% most unlucky people!\n")
        print("Cumulative probability is", round(cumulativeProbability*100,4), "%")
    if (autoFlag=='self'):
        print("Type 'go' to enhance, or 'exit' to quit")
        userIn=input()
print("Finished")
```
